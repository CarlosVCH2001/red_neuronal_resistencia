```python
#Importamos las librerías
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
# Limpiar la sesión de Keras (si se estaba utilizando previamente)
from tensorflow.keras import backend as K
K.clear_session()
```

```python
# Cargar datos desde un archivo Excel
data = pd.read_excel('../dataset/ms/dataset-ms.xlsx', sheet_name='train')
```

```python
# Extraer características de entrada y etiquetas de salida
X = data[[
    'Cemento', 'Agua', 'AgregadoFino', 'AgregadoGrueso', 'TiempoCurado'
]]

# Convertir los datos de entrada a tipo float32
X = X.astype('float32')

y = data['Carga']
```

```python
# Definir la arquitectura de la red
modelo = tf.keras.Sequential([
    tf.keras.layers.Dense(units=32, activation='relu', input_shape=(5,)),
    tf.keras.layers.Dense(units=1)
])
```

```python
# Compilar el modelo
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
    loss='mean_squared_error'
)
```

```python
# Entrenar el modelo con todos los datos
modelo.fit(X, y, epochs=6000, verbose=True)
```

```python
# Función para realizar una predicción
def predecir_carga():
    print("Ingresa los valores de las 6 características:")
    cemento = float(input("Cemento: "))
    agua = float(input("Agua: "))
    agregado_fino = float(input("AgregadoFino: "))
    agregado_grueso = float(input("AgregadoGrueso: "))
    tiempo_curado = float(input("TiempoCurado: "))

    entrada = np.array([cemento, agua, agregado_fino, agregado_grueso, tiempo_curado])
    
    carga_predicha = modelo.predict(np.array([entrada]))

    print(f"Carga predicha: {carga_predicha[0][0]}")  # No se especifica la unidad, asegúrate de que coincida con tus datos

# Llamar a la función para realizar una predicción
predecir_carga()
```
