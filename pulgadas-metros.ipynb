```python
#Importar librerias
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Declaramos los eventos pasados
entrada=np.array([1,6,30,7,70,43,503,201,1005,99],dtype=float)

resultados=np.array([0.0254, 0.1524, 0.762, 0.1778, 1.778, 1.0922, 12.776, 5.1054, 25.527, 2.514], dtype=float)
```

```python
#Topografía de la red
capa1 = tf.keras.layers.Dense(units=1, input_shape=[1])
```

```python
#Creamos el tpo de red
modelo = tf.keras.Sequential([capa1])
```

```python
# Asigna el optimizador y la métrica de pérdida
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.1),
    loss='mean_squared_error'
)

print("Entrenando la red")
```

```python
#Entrenamiento del modelo
entrenamiento=modelo.fit(entrada, resultados, epochs=500, verbose=True)
```

```python
#Guardar la red
modelo.save('modelo.h5')
```

```python
# Cargar el modelo previamente entrenado
modelo = tf.keras.models.load_model('modelo.h5')
```

```python
#Observar el comportamiento de nuetra red
plt.xlabel('Ciclos de entrenamiento')
plt.ylabel('Errores')
plt.plot(entrenamiento.history["loss"])
plt.show()
```

```python
#Prediccion
i=input("Ingresar el valor en pulgadas:")
i=float(i)

prediccion=modelo.predict([i])

print("El valor en pulgadas: ", i)
print("El valor en metros: ", str(prediccion))
```
